```python
import matplotlib.pyplot as plt

#Classe para representar um livro
class Livros:
  def __init__(self, titulo, autor, genero, quantidade_disponivel):
    self.titulo = titulo
    self.autor = autor
    self.genero = genero
    self.quantidade_disponivel = quantidade_disponivel

  def __str__(self):
    return f"{self.titulo} por {self.autor}, gênero: {self.genero}, obras disponíveis: {self.quantidade_disponivel}"

#Criar uma lista de livros
biblioteca = []

#Lista para armazenar genero
generos = []

#Função para adicionar um livro à biblioteca
def adicionar_livro(titulo,autor,genero, quantidade_disponivel=1):
  novo_livro = Livros(titulo,autor,genero, quantidade_disponivel)
  biblioteca.append(novo_livro)
  generos.append(genero)  #Adiciona o genero à lista de genero
  print(f'O livro "{titulo}" foi adicionado à biblioteca.')

#Função para listar todos os livros na biblioteca
def listar_livros():
  print('Livros na biblioteca: ')
  for livro in biblioteca:
    print(livro)

#Adicionar alguns livros à biliboteca
adicionar_livro('Dom Casmurro', 'Machado de Assis', 'Romance')
adicionar_livro('Orgulho e Preconceito', 'Jane Austin', 'Romance')
adicionar_livro('1984', 'George Orwell', 'Ficção Distópica')
adicionar_livro('Cem anos de solidão', 'Gabriel Garcia Marquez', 'Realismo Mágico')
adicionar_livro('Apanhados dno campo de centeio', 'J.D. Salinger', 'Ficção')
adicionar_livro('O Alquimista', 'Paulo Coelho', 'Romance')

#Listar todos os livros na biblioteca
listar_livros()

#Função para busca de um livro pelo seu título
def buscar_livro_por_titulo(titulo_busca):
  for livro in biblioteca:
    if livro.titulo.lower() == titulo_busca.lower():
      return livro
  return None

#Buscar livro pelo título
livro_busca = buscar_livro_por_titulo(input('Digite o título do livro que deseja buscar: '))
if livro_busca:
  print(f'Livro encontrado: {livro_busca}')
else:
  print('Livro não encontrado.')

#Criar um gráfico de livros por genero
generos_unicos = list(set(generos))  #Remove duplicata de gêneros
generos_unicos.sort()

# Contagem de livros por genero
contagem_por_genero = [generos.count(generos_unico) for generos_unico in generos_unicos]

#Criar um gráfico de linha
plt.plot(generos_unicos, contagem_por_genero, marker='o', linestyle='-')
plt.xlabel('Gênero literário')
plt.ylabel('Número de livros')
plt.title('Distribuição de livros na biblioteca por gênero literário')

#Adicionar rótulos aos pontos de dados
for i, valor in enumerate(contagem_por_genero):
  plt.text(generos_unicos[i], valor, str(valor), ha='center', va='bottom')

#Mostrar o gráfico
plt.show()
```
